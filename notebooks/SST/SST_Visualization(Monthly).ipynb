```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV data (replace with the actual path to your file)
data = pd.read_csv('Mean_SST_Values.csv')

# Check the first few rows
print(data.head())
```

```python
# Convert the 'date' column to datetime format
data['date'] = pd.to_datetime(data['date'], format='%Y-%m-%d')

# Extract the month from the 'date' column
data['month'] = data['date'].dt.month
```

```python
# Group by month and calculate the mean SST for each month
monthly_mean_sst = data.groupby('month')['mean_sst'].mean()
```

```python
# Plot the monthly mean SST values as a line graph
plt.figure(figsize=(10, 6))
plt.plot(monthly_mean_sst.index, monthly_mean_sst.values, marker='o', linestyle='-', color='skyblue', markersize=6)

# Customize plot labels and title
plt.title('Monthly Mean SST Values for 2023', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Mean SST', fontsize=12)
plt.xticks(monthly_mean_sst.index)  # Ensures all months (1 to 12) are shown on the x-axis
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```
